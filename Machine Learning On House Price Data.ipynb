### null value handling

```python
# find the number of missing records
df_house_price.isnull().sum()
```

```python
# N_Hospital_Beds,we have less than ten percent data missing
df_house_price.dropna(inplace=True)
```

```python
# Do the authentic check
df_house_price.isnull().sum()
```

```python
df_target=df_house_price['price']
df_feature=df_house_price.drop('price',axis=1)
```

```python
# segregate the data based on datatype
df_num=df_house_price.select_dtypes(include='number')
df_cat=df_house_price.select_dtypes(include='object')
```

```python
# Show the numerical data
df_num.head()
```

```python
df_cat.head()
```

### Perform encoding on categorical variable

```python
# use dummy encoding
df_encoded=pd.get_dummies(df_cat)
# reset the index space 
df_encoded=df_encoded.reset_index()
```

```python
# Show th encoded data
df_encoded.head()
```

### Perform Scaling On Numerical Data

```python
# Import the standard scaler
from sklearn.preprocessing import StandardScaler 

# initialte the standard scalar 
ss=StandardScaler()

# scale the value present in numerical data frame
scaled=ss.fit_transform(df_num)

# numerate the data into the dataframe
scaled_df=pd.DataFrame(scaled,columns=df_num.columns)

# reset the index
scaled_df=scaled_df.reset_index()
```

```python
# Show the data 
scaled_df.head()
```

### Perform the Concatenation Operation

```python
# concate encoded variables and state variables
df_independent=pd.concat([df_encoded,scaled_df],axis=1)
```

```python
#Show the data
df_independent.head()
```

## Split the Data for Traininga and Testing

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data 
x_train,x_test,y_train,y_test=train_test_split(df_independent,df_target,test_size=0.2)
```

```python
# check the size
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# import the model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Initialize the model
LR=LinearRegression()
```

```python
# frame the linear regression model
model_lr=LR.fit(x_train,y_train)
```

```python
# perform the prediction
predicted=model_lr.predict(x_test)
```

```python
# show the predictive model
predicted
```

```python
y_test
```

```python
# evaluate the model
round(mean_squared_error(predicted,y_test),3)
```
